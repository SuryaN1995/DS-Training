```python
import numpy as np
import pandas as pd
%matplotlib inline
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
```

```python
X = 2.5* np.random.randn(100)+1.5
res = 0.5* np.random.randn(100)
Y = 2+ 0.3*X +res
```

```python
df = pd.DataFrame({'X':X,'Y':Y})
```

```python
df
```

```python
xmean = np.mean(X)
```

```python
ymean = np.mean(Y)
df['xycon'] = (df['X'] - xmean) * (df['Y'] - ymean)
df['xvar'] = (df['X'] -xmean)**2
```

```python
beta = df['xycon'].sum()/df['xvar'].sum()
```

```python
alpha = ymean - (beta - xmean)
```

```python
ypred = alpha +beta+X
```

```python
plt.figure(figsize=(12,6))
```

```python
plt.plot(X,ypred)
```

```python
plt.plot(X,Y,'ro')
```

```python
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# sklearn linear
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('adv.csv')
```

```python
df.head()
```

```python
cols = ['TV','Radio']
x1 = df[cols]
y1 = df['Sales']
```

```python
lr = LinearRegression()
```

```python
model = lr.fit(x1,y1)
```

```python
model
```

```python
print('beta = {}'.format(model.coef_))
print('alpha = {}'.format(model.intercept_))
```

```python
model.predict(x1)
```
