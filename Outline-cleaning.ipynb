```python
lower_bound = 0.05
upper_bound = 0.96
res = df['Values'].quantile([lower_bound,upper_bound])
```

```python
res
```

```python
true_index = (res.loc[lower_bound]<df.Values.values) & (res.loc[upper_bound]>df.Values.values)
```

```python
true_index
```

```python
false_index =  true_index != True
```

```python
false_index
```

```python
df.Values[true_index]
```

```python
df.Values[false_index]
```

```python
np.median(df.Values)
```

```python
df.Values[false_index] = np.median(df.Values)
```
