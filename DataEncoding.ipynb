```python
def encoder(d,c):
    le = preprocessing.LabelEncoder()
    le.fit(d[c])
    after_Trans = le.transform(d[c])
    d[c] = after_Trans
    return d
```

```python
result = encoder(data,'OWN_OCC')
```

```python
# Hot Encoding
data = pd.read_csv('cleaned.csv')
```

```python
data
```

```python
one_dummies = pd.get_dummies(data['OWN_OCC'],prefix='OWN_OCC')
```

```python
one_dummies
```

```python
data.drop('OWN_OCC',inplace=True,axis=1)
```

```python
data
```

```python
data = data.join(one_dummies)
```

```python
data
```

```python
def hotEncoding(d,c):
    one_dummies = pd.get_dummies(d[c],prefix=c)
    d.drop(c,inplace=True,axis=1)
    d = d.join(one_dummies)
    return d
```

```python
hotEncoding(data,"ST_NAME")
```
