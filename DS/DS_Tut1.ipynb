```python
#Encoding
def one_hot_en(d,c):
    one_dummies = pd.get_dummies(d[c],prefix=c)
    d.drop(c,inplace=True,axis=1)
    d = d.join(one_dummies)
    return d
```

```python
df = one_hot_en(df,'Gender')
```

```python
df.head()
```

```python
df = one_hot_en(df,'Married')
df = one_hot_en(df,'Education')
df = one_hot_en(df,'Self_Employed')
df = one_hot_en(df,'Loan_Status')
df = one_hot_en(df,'Property_Area')
```

```python
df.drop('Loan_ID',axis=1,inplace=True)
```
